```python
import pandas as pd
import numpy as np
```

```python
tennis = pd.read_csv("tennis.csv")
wine=pd.read_csv("WineQT.csv")
books = pd.read_csv("nyt_full.tsv", sep="\t")
```

```python
tennis.head()
```

```python
tennis.drop(['Tournament','Rank'], inplace=True, axis=1)
```

```python
tennis.columns
```

```python
tennis.columns=['event','runner-up','score','winner','year']
```

```python
tennis.head()
```

```python
def set_event_order(n):
    events={'Australian Open':1,
            'Roland Garros':2,
            'Wimbledon':3,
            'US Open':4}
    return events[n]
```

```python
set_event_order('Wimbledon')
```

```python
tennis['order']=list(map(lambda x: set_event_order(x), tennis['event']))
```

```python
tennis.head()
```

```python
tennis=tennis[['event','order','runner-up','score','winner','year']]
```

```python
tennis.head()
```

```python
tennis['winner'].groupby(tennis['winner']).count().sort_values(ascending=False)
```

```python
federer=tennis.query('winner=="Roger Federer"')
```

```python
federer.head()
```

```python
federer['event'].groupby(federer['event']).count()
```
